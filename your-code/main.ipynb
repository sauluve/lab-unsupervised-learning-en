```python
# Your code here
# customers_transformed = data
# customers_transformed["Store-ID"] = customers.groupby(["Channel","Region"]).ngroup()
# # customers_transformed = customers_transformed.drop(columns=["Channel","Region"], axis=1)
# # customers_transformed = customers_transformed.groupby("Store-ID").agg("sum")
# customers_transformed_feats = customers_transformed[["Fresh","Milk","Grocery","Frozen","Detergents_Paper","Delicassen"]]
# customers_transformed_labels = customers_transformed[["Channel","Region","Store-ID"]]
# customers_transformed

# Drop "Channel" and "Region"
customers = customers.drop(columns=["Channel","Region"], axis=1)
```

```python
# Drop the outliers
def drop_outliers(df):
    # Find the upper limit of the whisker (in this case, there's no need for a lower limit)
    df_describe = df.describe().T
    df_describe["upper"] = (df_describe["75%"] - df_describe["25%"])*1.5 + df_describe["75%"]

    # If any row has a column with a value higher than the corresponding upper limit for that column, that value is deleted (turned into NaN)
    # df_filtered = df[df > df_describe["upper"]]
    df_filtered = df.copy()

    for feature in df.columns:
        upper_bound = df_describe.loc[feature,"upper"]
        df_filtered = df_filtered[df_filtered[feature] <= upper_bound]

    # Return only the rows without any Nan
    return df_filtered.dropna(axis=0)

customers_without_outliers = drop_outliers(customers)
plt.boxplot(customers_without_outliers)
plt.show()
plt.boxplot(customers);
```

# Challenge 3 - Data Preprocessing

One problem with the dataset is the value ranges are remarkably different across various categories (e.g. `Fresh` and `Grocery` compared to `Detergents_Paper` and `Delicassen`). If you made this observation in the first challenge, you've done a great job! Keep on the good work!

Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

```python
# Your import here:

from sklearn.preprocessing import StandardScaler

# Your code here:
scaler = StandardScaler()

customers_scale = scaler.fit_transform(customers)
display(customers_scale)
```

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
from sklearn.cluster import KMeans

# Your code here:
kmeans_model = KMeans()

customers["Labels_KMeans"] = kmeans_model.fit(customers_scale).labels_
display(customers)
```

### Looking to the elbow we can choose 2 like the correct number of clusters

```python
kmeans_2 = KMeans(n_clusters=2).fit(customers_scale)

labels = kmeans_2.predict(customers_scale)

clusters = kmeans_2.labels_.tolist()
```

```python
clean_customers = customers
clean_customers['Label'] = clusters
clean_customers
```

Count the values in `labels`.

```python
# Your code here:
clean_customers["Labels_KMeans"].value_counts()
```
